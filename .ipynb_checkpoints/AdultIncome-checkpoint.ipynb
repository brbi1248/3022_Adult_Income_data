## Data Transformations: Encode All Attributes As Numeric
 

For easy interpretability, I decided to encode my target variable "Income" from <=50K to 0 and >50K to 1.

```python
income_df['Income'] = income_df['Income'].apply(lambda x: 0 if x.strip() == '<=50K' else 1)
```

```python
income_df['Income'].value_counts()
```

Reference the category numbers:

All categorical variables in the dataset were converted into numeric variables for easier plotting and data analysis. The categories are printed below with their new numbers so plots can be easily interpreted. Note that Age, Education-Num, and Hours per Week did not need to be transformed because they were already numeric.

```python
categorical_vars = ['Race', 'Occupation', 'MaritalStatus', 'Sex', 'NativeCountry']

category_mapping = {}

# Encode categorical variables as numeric in income_df and store the mappings
for var in categorical_vars:
    label_encoder = LabelEncoder()
    income_df[var] = label_encoder.fit_transform(income_df[var])
    category_mapping[var] = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

for var, mapping in category_mapping.items():
    print(f"{var} mapping:")
    for category, encoded_category in mapping.items():
        print(f"{category} : {encoded_category}")
    print()
```

## Logistic Regression Model

A logistic regression model is an appropriate choice for this dataset because the target variable is a binary classification of income groups. While there is some mild collinearity between features, logistic regression can handle this effectively. I will attempt to address the data imbalance by using over-sampling and under-sampling.

```python
# logistic function 
def logistic(model, x):
    z = np.exp(model.intercept_ + model.coef_[0] * x)
    return  z / (1 + z)
```

```python
# Instantiate a logistic regression model that has default parameters and fit with training set
lr = sklearn.linear_model.LogisticRegression().fit(X_train, y_train)

# Predict instances from the test set
y_hat = lr.predict(X_test)

print("Accuracy:")
print(classification_report(y_test, y_hat))
```

## Improving Logistic Regression Model

### Random Undersampling

```python
rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X_train, y_train)

# Instantiate a logistic regression model and fit with random undersampling training set
lr = LogisticRegression().fit(X_resampled, y_resampled)

# Predict using the model
y_hat = lr.predict(X_test)
conf_matrix = metrics.confusion_matrix(y_test, y_hat)
accuracy = accuracy_score(y_test, y_hat)

print("Confusion Matrix:")
print(conf_matrix)
print("Accuracy:")
print(accuracy)
```

### Random Oversampling

```python
# Using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Instantiate a logistic regression model and fit with random oversampling training set
lr = LogisticRegression().fit(X_resampled, y_resampled)

# Predict using the model
y_hat = lr.predict(X_test)

conf_matrix = metrics.confusion_matrix(y_test, y_hat)
accuracy = accuracy_score(y_test, y_hat)

print("Confusion Matrix:")
print(conf_matrix)
print("Accuracy:")
print(accuracy)
```

### HyperTuning

```python
# Define the grid values
tol = [0.01, 0.001, 0.0001]
max_iter = [100, 150, 200]

# Create a dictionary of tol and max_iter
param_grid = dict(tol=tol, max_iter=max_iter)

# Instantiate GridSearchCV 
grid_model = GridSearchCV(estimator=lr, param_grid=param_grid, cv=5)
scaler = StandardScaler()

# Use scaler to rescale X 
rescaledX = scaler.fit_transform(X)

# Fit data to grid_model
grid_model_result = grid_model.fit(rescaledX, y)

best_score, best_params = grid_model_result.best_score_, grid_model_result.best_params_
print("Accuracy:")
print(best_score, best_params)
```

Based on the three methods I chose to improve the model's accuracy it looks like hypertuning helped improved the score, but only by 1%. Even so, I will be using this going forward with feature analysis and selection to further improve my model.

```python
# *This may take a couple minutes to run as it is iterating through every combination of predictor interactions*
# Initialize results dictionary
results = {}

# Generate all feature combinations
for num_features in range(1, len(predictors) + 1):
    for combo in combinations(predictors, num_features):
        feature_formula = '+'.join(combo)
        formula = f'Income ~ {feature_formula}'
        
        # Train RandomForestClassifier
        model = RandomForestClassifier(random_state=42)
        model.fit(X_train[list(combo)], y_train)
        
        # Predict test data
        y_pred = model.predict(X_test[list(combo)])
        
        accuracy = accuracy_score(y_test, y_pred)
        results[feature_formula] = accuracy

sorted_results = sorted(results.items(), key=lambda x: x[1], reverse=True)
for feature_formula, accuracy in sorted_results:
    print(f"Feature Formula: {feature_formula}, Accuracy: {accuracy:.4f}")
```
